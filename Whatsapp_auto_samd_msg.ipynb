```python
import pywhatkit
import pyautogui
import time
from datetime import datetime, timedelta

def send_whatsapp_now(phone_number, message):
    try:
        # Current time se 2 minute baad ka time set karo
        now = datetime.now() + timedelta(minutes=2)
        hour = now.hour
        minute = now.minute
        
        print(f"Message will be sent at {hour}:{minute}")
        print("Do not touch mouse/keyboard until message is sent!")
        
        # Message send karo
        pywhatkit.sendwhatmsg(phone_number, message, hour, minute, wait_time=20, tab_close=False)
        
        # Message type hone ka wait karo
        time.sleep(3)
        
        # Enter press karo
        pyautogui.press('enter')
        
        print("✅ Message sent successfully!")
        
        # Browser tab close karo
        time.sleep(2)
        pyautogui.hotkey('ctrl', 'w')
        
    except Exception as e:
        print(f"❌ Error occurred: {e}")

# Usage - Message abhi send hoga
phone_number = "+923136503112"
message = "Hello! Ye message automatically Python se send hua hai! 🚀"

send_whatsapp_now(phone_number, message)
```

```python
import subprocess
import time
import urllib.parse

def send_whatsapp_via_app(phone_number, message):
    """
    Android WhatsApp app me message send karta hai ADB ke through
    Requirements: 
    1. Android phone USB se connected ho
    2. USB Debugging enabled ho
    3. ADB installed ho computer me
    """
    
    try:
        # Phone number format (remove +)
        if phone_number.startswith('+'):
            phone_number = phone_number[1:]
        
        # Message encode karo URL ke liye
        encoded_message = urllib.parse.quote(message)
        
        # WhatsApp URL create karo
        whatsapp_url = f"https://api.whatsapp.com/send?phone={phone_number}&text={encoded_message}"
        
        print("📱 Opening WhatsApp on phone...")
        
        # ADB command se phone me WhatsApp open karo
        cmd = f'adb shell am start -a android.intent.action.VIEW -d "{whatsapp_url}"'
        subprocess.run(cmd, shell=True, check=True)
        
        print("⏳ Waiting for WhatsApp to load...")
        time.sleep(5)
        
        # Send button pe tap karo (coordinates adjust karne padenge)
        # Ye coordinates Samsung Galaxy ke liye hai - aap ke phone ke according change karo
        print("📤 Sending message...")
        subprocess.run("adb shell input tap 950 1850", shell=True)  # Send button coordinates
        
        print("✅ Message sent successfully via WhatsApp App!")
        
    except subprocess.CalledProcessError as e:
        print(f"❌ ADB Error: {e}")
        print("💡 Make sure:")
        print("   - Phone is connected via USB")
        print("   - USB Debugging is enabled")
        print("   - ADB is installed")
        
    except Exception as e:
        print(f"❌ Error: {e}")

# Usage
phone_number = "+923218636730"
message = "Hello! Ye message WhatsApp app se directly aya hai! 📱"

send_whatsapp_via_app(phone_number, message)
```
